# 範囲の選択と抽出

### PLATEAUデータの読み込み

`load_dataset()` にIDを指定してデータセットを読み込みます。

```python
from plateaukit import load_dataset

tokyo23ku = load_dataset("plateau-tokyo23ku")

tokyo23ku
```

### 範囲の選択

`.get_area()` メソッドを使って範囲を選択します。

- 引数は `[左上のx座標 (経度), 左上のy座標 (緯度), 右下のx座標 (経度), 右下のy座標 (緯度)]` で指定します。
- 引数を指定しない場合は全範囲が選択されます。

```python
area = tokyo23ku.get_area([139.7596, 35.6858, 139.7724, 35.6752])
```

### 範囲のプレビュー

`.show()` メソッドを使って範囲をプレビューできます。

```python
area.show()
```

### 範囲のデータを取得

`.gdf` プロパティから範囲のデータ (`geopandas.GeoDataFrame`) を取得できます。

```python
area.gdf
```

### GeoJSON への変換

`.to_geojson()` メソッドを使って選択範囲を GeoJSON 形式に変換できます。

```python
data = area.to_geojson()

data[:300]
```
